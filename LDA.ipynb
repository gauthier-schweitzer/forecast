### Data recovery

In the scraping data, we have not included the id of the thread, we would like to access this information. We will therefore try to recover it.

```python
data["datetime"]=pd.to_datetime(data["datetime"])
data['datetime_lag']=data['datetime'].shift(1)
#pd.to_datetime(data["datetime_lag"])
data.head()
```

```python
data['thread_id'] = data['datetime']<data['datetime_lag']

index = 0

def thread_mapping (row):
    global index
    if row['thread_id']:
        index +=1
        temporal = index
        return(temporal)
    else: 
        temporal = index
        return(temporal)
    
thread_id = data.apply(thread_mapping, axis=1)
data['thread_id']=thread_id
```

```python
post_by_thread = data.groupby('thread_id')['post'].apply(lambda x: "{%s}" % ', '.join(x))
```

```python
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
import gensim
import progressbar

tokenizer = RegexpTokenizer(r'\w+')

# create English stop words list
en_stop = get_stop_words('en')

# Create p_stemmer of class PorterStemmer
p_stemmer = PorterStemmer()
    
# list for tokenized documents in loop
texts = []

# loop through document list
for i in progressbar.progressbar(post_by_thread):
    
    # clean and tokenize document string
    raw = i.lower()
    tokens = tokenizer.tokenize(raw)

    # remove stop words from tokens
    stopped_tokens = [i for i in tokens if not i in en_stop]
    
    # stem tokens
    stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
    
    # add tokens to list
    texts.append(stemmed_tokens)

# turn our tokenized documents into a id <-> term dictionary
dictionary = corpora.Dictionary(texts)
    
# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
# generate LDA model
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=50, id2word = dictionary, passes=20)
```
